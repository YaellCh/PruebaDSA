## Explicacion de que es una lista enlazada simple

1.3 Resumen de listas enlazadas 

- Están compuestas por nodos.
- Los nodos contienen un enlace al siguiente nodo (y también al nodo anterior en el caso de listas enlazadas bidireccionales o doblemente enlazadas).
- Pueden ser unidireccionales o bidireccionales.
- Son una estructura de datos básica y forman la base de muchas otras estructuras de datos
- Tienen un único nodo head, que actúa como el primer nodo de la lista.
- Requieren cierto mantenimiento para agregar o eliminar nodos.
- Su implementacion depende del caso de uso exacto o del lenguaje de programación que se utilice.

## Speech del video 

Hola, bienvenidos a todos. Hoy hablaremos brevemente sobre **colas en Estructuras de datos en python** pero antes de empezar con ese tema debemos conocer antes algunos conceptos que son claves para una mejor comprension del tema. Comenzamos con las listas enlazadas.

- **¿Qué son?**
  
Las listas enlazadas están formadas por nodos, pequeñas unidades que contienen dos cosas: el dato que almacenan y un enlace al siguiente nodo. En el caso de las listas doblemente enlazadas, también tienen un enlace al nodo anterior.

Tipos de listas enlazadas:
Pueden ser unidireccionales, donde cada nodo solo apunta hacia adelante, o bidireccionales, que permiten moverse en ambas direcciones.

- **¿Por qué son importantes?**
  
Estas listas son la base de muchas estructuras de datos avanzadas. Además, cuentan con un único nodo inicial llamado head, que actúa como el punto de partida de la lista.

- **¿Qué implica trabajar con ellas?**

Modificar una lista enlazada, ya sea agregando o eliminando nodos, requiere cierta atención, ya que los enlaces entre nodos deben mantenerse correctamente. Su implementación también puede variar dependiendo del lenguaje de programación y las necesidades específicas del proyecto.

En resumen, las listas enlazadas son herramientas poderosas para organizar y gestionar datos de forma eficiente.



# Video numero 2 

## Listas doblemente enlazadas:


- Las listas doblemente enlazadas están compuestas por nodos que contienen enlaces a los nodos siguientes y anteriores.
- Son bidireccionales, lo que significa que se pueden recorrer en ambas direcciones.
- Tienen un puntero a un solo nodo _cabecera_, que sirve como el primer nodo de la lista.
- Tienen un puntero a un solo nodo _cola_, que sirve como el último nodo de la lista.
- Requieren que los punteros de la cabecera de la lista se actualicen después de agregar o quitar el `head_node`.
- Requieren que los punteros del final de la  lista se actualicen después de agregar o quitar el  `tail_node`.
- Requieren que los punteros de los nodos circundantes se actualicen después de quitar un nodo de la parte media de la lista.

El historial de tu navegador es un ejemplo de una lista doblemente enlazada. Cuando abres tu navegador, la primer página a la que accesas se guarda es la cabecera de la lista. A medida que haces clic en las cosas y navegas a nuevas páginas, estas se agregan a la parte  inferior de la lista. Si alguna vez deseas volver a algo que ya has visitado, puedes usar el botón “atrás” para retroceder en la lista.

# Speech del video:

Continuamos explorando las listas doblemente enlazadas, una estructura de datos clave.

- **¿Qué son?**

Las listas doblemente enlazadas están formadas por nodos, pero a diferencia de las unidireccionales, cada nodo tiene enlaces a los nodos siguientes y anteriores.

- **Características principales:**

Son completamente bidireccionales, lo que significa que puedes recorrerlas en ambas direcciones.
Cuentan con un nodo inicial llamado cabecera (head), que es el primer nodo de la lista, y un nodo final llamado cola (tail), que es el último.
Para agregar o eliminar un nodo en cualquier posición, es necesario actualizar cuidadosamente los enlaces de los nodos circundantes y, en algunos casos, los punteros de la cabecera o la cola.

- **¿Por qué son útiles?**

Un gran ejemplo es el historial de tu navegador web. Imagina que la primera página que visitas es la cabecera de la lista. A medida que navegas, las nuevas páginas se añaden al final. Si deseas retroceder, puedes usar el botón "atrás" para moverte en la lista hasta una página anterior.

En resumen, las listas doblemente enlazadas son versátiles y perfectas para situaciones donde necesitas navegar datos en ambas direcciones o mantener flexibilidad en las modificaciones.

1. Introduccion a los Nodos

La primera estructura de datos que abordarás son los nodos. Aprenderás qué son y cómo implementarlos en Python. Los nodos son los componentes básicos de muchas de las estructuras de datos que siguen y los verás aparecer muchas veces a lo largo del curso.

Un nodo individual contiene datos y enlaces a otros nodos. Considera el nodo representado en la siguiente imagen. Este nodo `(node_a)` contiene un fragmento de datos (el número 5) y un vínculo o enlace a otro nodo `(node_b)`.

Los enlaces dentro del nodo a veces se denominan **punteros**, ya que “apuntan” a otro nodo.

Normalmente, las estructuras de datos implementan nodos con uno o más enlaces. Si estos enlaces son nulos (`null`), significa que ha llegado al final del nodo o ruta de enlace particular que estaba siguiendo anteriormente.

En la siguiente imagen se muestran diversas implementaciones de nodos. Examina los tipos de datos y cómo se vinculan algunos de los nodos.


## 1.1 Nodo huérfano 

En el contexto de las estructuras de datos, un **nodo huérfano** se refiere generalmente a un nodo que ha perdido su nodo padre o que ya no está conectado a la estructura principal en la que residía. Esto puede ocurrir en diferentes contextos:

1. **Estructuras de Árbol:** En una estructura de árbol, un nodo se convierte en huérfano si pierde su nodo padre sin ser reinsertado adecuadamente en otra parte del árbol. Por ejemplo, si el nodo padre es eliminado o movido sin volver a conectar el nodo hijo, este último se convierte en un nodo huérfano.

2. **Estructuras de Grafo:** En un grafo, un nodo huérfano puede referirse a un nodo que ya no está conectado a ningún otro nodo, quedando efectivamente aislado del resto del grafo.

3. **Listas Enlazadas:** En el contexto de las listas enlazadas, un nodo huérfano puede ocurrir si un nodo se elimina de la lista y sus referencias (enlaces al siguiente o anterior nodo) no se actualizan correctamente, dejando el nodo en la memoria pero ya no formando parte de la lista.

Los nodos huérfanos suelen ser indeseables porque pueden llevar a pérdidas de memoria o inconsistencias en la estructura de datos, ya que estos nodos ya no forman parte de la estructura prevista pero siguen ocupando recursos.


## 1.2 Resumen de Nodos

Los nodos:
- Contienen datos, que pueden ser de distintos tipos.
- Contienen vínculos a otros nodos. Si un nodo no tiene vínculos, o todos sus vinculos son nulos, has llegado al final de la ruta que estabas siguiendo.
- Pueden quedar huérfanos si no existen vínculos a ellos.


# Speech del video:

- **¿Qué es un nodo?**
  
Un nodo es una unidad básica que contiene:
Datos, que pueden ser números, texto o cualquier tipo de información.
Enlaces o punteros que conectan ese nodo con otros, formando estructuras como listas, árboles o grafos.
Imagina un nodo llamado node_a. Este puede almacenar un dato, como el número 5, y tener un enlace que apunta a otro nodo, llamado node_b. Si el enlace está vacío o es null, significa que has llegado al final de la ruta.

- **¿Qué pasa con los nodos huérfanos?**

Un nodo huérfano es un nodo que ha perdido su conexión con la estructura principal, lo que puede ocurrir en diferentes escenarios:
En árboles, si un nodo pierde su padre.
En grafos, cuando un nodo queda aislado.
En listas enlazadas, si un nodo se elimina incorrectamente y sus referencias no se actualizan.
Esto puede generar problemas como pérdidas de memoria o errores en el manejo de la estructura.

**Resumen:**
Los nodos son esenciales porque:
Contienen los datos que manejamos.
Forman enlaces que conectan estructuras.
Pueden quedar huérfanos si no se gestionan correctamente.
