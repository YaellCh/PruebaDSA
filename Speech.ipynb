## Listas doblemente enlazadas:


- Las listas doblemente enlazadas están compuestas por nodos que contienen enlaces a los nodos siguientes y anteriores.
- Son bidireccionales, lo que significa que se pueden recorrer en ambas direcciones.
- Tienen un puntero a un solo nodo _cabecera_, que sirve como el primer nodo de la lista.
- Tienen un puntero a un solo nodo _cola_, que sirve como el último nodo de la lista.
- Requieren que los punteros de la cabecera de la lista se actualicen después de agregar o quitar el `head_node`.
- Requieren que los punteros del final de la  lista se actualicen después de agregar o quitar el  `tail_node`.
- Requieren que los punteros de los nodos circundantes se actualicen después de quitar un nodo de la parte media de la lista.

El historial de tu navegador es un ejemplo de una lista doblemente enlazada. Cuando abres tu navegador, la primer página a la que accesas se guarda es la cabecera de la lista. A medida que haces clic en las cosas y navegas a nuevas páginas, estas se agregan a la parte  inferior de la lista. Si alguna vez deseas volver a algo que ya has visitado, puedes usar el botón “atrás” para retroceder en la lista.

## 1.1 Nodo huérfano 

En el contexto de las estructuras de datos, un **nodo huérfano** se refiere generalmente a un nodo que ha perdido su nodo padre o que ya no está conectado a la estructura principal en la que residía. Esto puede ocurrir en diferentes contextos:

1. **Estructuras de Árbol:** En una estructura de árbol, un nodo se convierte en huérfano si pierde su nodo padre sin ser reinsertado adecuadamente en otra parte del árbol. Por ejemplo, si el nodo padre es eliminado o movido sin volver a conectar el nodo hijo, este último se convierte en un nodo huérfano.

2. **Estructuras de Grafo:** En un grafo, un nodo huérfano puede referirse a un nodo que ya no está conectado a ningún otro nodo, quedando efectivamente aislado del resto del grafo.

3. **Listas Enlazadas:** En el contexto de las listas enlazadas, un nodo huérfano puede ocurrir si un nodo se elimina de la lista y sus referencias (enlaces al siguiente o anterior nodo) no se actualizan correctamente, dejando el nodo en la memoria pero ya no formando parte de la lista.

Los nodos huérfanos suelen ser indeseables porque pueden llevar a pérdidas de memoria o inconsistencias en la estructura de datos, ya que estos nodos ya no forman parte de la estructura prevista pero siguen ocupando recursos.


## 1.2 Resumen de Nodos

Los nodos:
- Contienen datos, que pueden ser de distintos tipos.
- Contienen vínculos a otros nodos. Si un nodo no tiene vínculos, o todos sus vinculos son nulos, has llegado al final de la ruta que estabas siguiendo.
- Pueden quedar huérfanos si no existen vínculos a ellos.

## Speech a cargo de Yael

¡Hola! Hoy vamos a hablar de una estructura de datos muy común y práctica: las colas.

- **¿Qué es una cola?**
  
Una cola es una estructura de datos ordenada que sigue el principio FIFO: First In, First Out, o lo que entra primero, sale primero. Es como una fila para comprar entradas de cine.
Las colas permiten tres operaciones básicas:

Encolar: agregar datos al final de la cola.
Desencolar: eliminar datos del frente de la cola.
Mirar: consultar el dato al frente de la cola sin eliminarlo.
¿Cómo se implementan?
Una cola puede implementarse usando una lista enlazada:
El frente de la cola corresponde al nodo cabeza.
El final de la cola corresponde al nodo cola.
Las modificaciones solo afectan los extremos, ya que no se permite intervenir en los nodos intermedios. Además, para mantener el acceso eficiente, es importante tener referencias tanto al frente como al final de la cola.

- **Colas limitadas y errores comunes:**

Si una cola tiene un límite de tamaño, se llama cola limitada. Aquí hay dos problemas que puedes encontrar:
Desbordamiento (Overflow): ocurre si intentas agregar datos a una cola llena.
Subdesbordamiento (Underflow): ocurre si intentas quitar datos de una cola vacía.
Ejemplo práctico:
Piensa en un cine. La primera persona en llegar a la fila será la primera en comprar su boleto. Si alguien intenta entrar en una fila que ya está llena, no podrá hacerlo (desbordamiento). Y si no hay nadie en la fila y alguien intenta ser atendido, no hay datos disponibles (subdesbordamiento).

**Resumen:**
Las colas admiten tres operaciones: encolar, desencolar y mirar.
Siguen una estructura FIFO.
Se pueden implementar con listas enlazadas.
Pueden tener límites de tamaño y ser susceptibles a desbordamientos o subdesbordamientos.

```python
## EJemplo de colas 

from Node_sandbox import Node

import logging
#Commit 3 no comentarios pk estan en los anteriores jjejeje

class Queue:
    def __init__(self, head_node=None, tail_node=None, size=0, max_size=None):
        self.head_node = head_node
        self.tail_node = tail_node
        self.size = size
        self.max_size = max_size

    def get_size(self):
        return self.size

    def has_space(self):
        if(self.max_size is None):
            return True
        else:
            if(self.max_size > self.get_size()):
                return True
            else:
                return False

    def is_empty(self):
        return self.size == 0
    
    def peek(self):
        if(self.is_empty()):
            print("¡No hay nada que ver aquí!")
        else:
            return self.head_node.get_value()
    
    def enqueue(self, value):
        if(self.has_space()):
            item_to_add = Node(value)
            print(f"¡Agregando {item_to_add.get_value()} a la cola!")
            if(self.is_empty()):
                self.head_node = item_to_add
                self.tail_node = item_to_add
            else:
                self.tail_node.set_next_node(item_to_add)
                self.tail_node = item_to_add
            self.size += 1
        else:
            logging.warning("¡Lo sentimos, no hay más espacio!")

    def dequeue(self):
        item_to_remove = Node(None)
        if(not self.is_empty()):
            item_to_remove = self.head_node
            logging.info("¡Eliminando" + str(item_to_remove.get_value()) + "de la cola!")
            if(self.get_size() == 1):
                self.head_node = None
            else:
                self.head_node = self.head_node.get_next_node()
            self.size = self.size - 1
        else:
            logging.info("¡Esta cola está totalmente vacía!")
        return item_to_remove.get_value()
```
